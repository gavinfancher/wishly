```python
import vobject

with open("contacts.vcf", "r") as f:
    data = f.read()

for card in vobject.readComponents(data):
    name = card.fn.value if "fn" in card.contents else "(No Name)"
    bday = card.bday.value if "bday" in card.contents else None
    if bday is not None:
        print(name, bday, sep=' -- ')
```
